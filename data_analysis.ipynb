```python
import os, json

path = './json_data/'

with os.scandir(path) as entries:
    files = [each.name for each in entries if 'corrupt' not in each.name and 'old' not in each.name]
files.pop(files.index('.DS_Store'))
print(len(files))
```

```python

def flatten(nested):
    if nested == []: return nested
    
    if isinstance(nested[0], list):
        return flatten(nested[0] + flatten(nested[1:]))
    return nested[:1] + flatten(nested[1:])

def generate_attributes(file, attribute):
    with open(f'{path}{file}') as jsonFile:
        nested_list = [[photos[each] for each in photos.keys() if each == attribute]\
                       for photos in json.load(jsonFile)['photos']['photo']]
        yield flatten(nested_list)

def get_attribute_data(files, attribute, num=0):
    '''Returns a list of dicts containing attribute values for each file in list\
    num is the slicing index for the files list (default is 0 which uses the entire file list)'''
    
    if num == 0:
        return ({file: [each for each in generate_attributes(file, attribute)]} for file in files)
    else:
        return ({file: [each for each in generate_attributes(file, attribute)]} for file in files[:num-1])
```
